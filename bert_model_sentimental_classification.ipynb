```python
import torch

# If there's a GPU available...
if torch.cuda.is_available():    

    # Tell PyTorch to use the GPU.    
    device = torch.device("cuda")

    print('There are %d GPU(s) available.' % torch.cuda.device_count())

    print('We will use the GPU:', torch.cuda.get_device_name(0))

# If not...
else:
    print('No GPU available, using the CPU instead.')
    device = torch.device("cpu")
```

```python
data=pd.read_csv('merged_sentimental/sentiment_final.csv')
```

```python
data['analytical','anger','confident','fear','sadness','tentative'] = data['analytical','anger','confident','fear','sadness','tentative'].fillna(0)
```

```python
data['analytical'] = data['analytical'].fillna(0)
```

```python
data['anger'] = data['anger'].fillna(0)
```

```python
data['confident'] = data['confident'].fillna(0)
```

```python
data['fear'] = data['fear'].fillna(0)
```

```python
data['sadness'] = data['sadness'].fillna(0)
```

```python
data['tentative'] = data['tentative'].fillna(0)
```

```python
data.head()
```

```python
data['labels']="default"
```

```python
data["labels"][(data["analytical"] == 1.0)] = "analytical"
data["labels"][(data["anger"] == 1.0)] = "anger"
data["labels"][(data["confident"] == 1.0)] = "confident"
data["labels"][(data["fear"] == 1.0)] = "fear"
data["labels"][(data["sadness"] == 1.0)] = "sadness"
data["labels"][(data["tentative"] == 1.0)] = "tentative"



```

```python
data['labels'].count()
```

```python
labeled_data=data[~data.labels.str.contains("default")]
```

```python
labeled_data=labeled_data.reset_index(drop=True)
```

```python
labeled_data.head()
```

```python
labeled_data.to_csv('merged_sentimental/sentiment_labeled_final.csv',index=False,encoding='utf8')
```
